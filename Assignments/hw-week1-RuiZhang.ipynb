**In a new notebook, copy the "student database"**

```python
student_database = [
  {'name':'john doe','id':12345,'courses':['EMSE6574','ENG101']},
  {'name':'jane doe','id':34567,'courses':['ENG101']},
  {'name':'moe doe','id':45678,'courses':['ENG101']},
]
```

**1 Write a function to register a new student**

* **The function should take a name and ID number**

```python
def register_student(name,id):
    student_database.append({'name':name,'id':id,'courses':[]})    ## Assigns an empty list to 'courses' to make it convenient to add classes later
```

```python
### Test case
register_student('bob doe',87654)
print(student_database)
```

**2 Write a function to find and print a student record by name**

```python
def find_record(name):
  found = False       ## "found" is used to flag if the name is found in the database
  for d in student_database:
    if d['name']== name:
      found = True     ## If typed in a correct name, set found to true
      print(d)
  if found == False:          ## If typed in a wrong name, will print warning
    print('Record not found')
```

```python
### Test case1: correct name
find_record('bob doe')    
```

```python
### Test case2: wrong name
find_record('tom')     
```

**3 Write a function to register a student for class**
* **The function should take ID number and class number**

```python
def register_class(id,class_number):
  for d in student_database:
    if d['id']== id:
      if class_number in d['courses']:         ## Checks if the class number is already registered
        print('Warning: The class was already registered!')    ## If the class number is already registered, will print warning
      else:                                    ## class number is not registered
        d['courses'].append(class_number)
        print('The class is successfully registered!')
```

```python
### Test case 1: id87654 has not registered ENG101 before
register_class(87654,'EMSE6574')  
print(student_database)
```

```python
### Test case 2: id12345 has already registered EMSE6754
register_class(12345,'ENG101')   
print(student_database)
```

* **Adding a logic to check if the id is in the database**

```python
def register_class2(id, class_number):
  FOUND = False    ## "FOUND" is used to check if the id is in the database
  for d in student_database:
    if d['id'] == id:
      FOUND = True 
      if class_number in d['courses']:
        print('Warning: The class was already registered!')    
      else:                                    
        d['courses'].append(class_number)
        print('The class is successfully registered!')
  if FOUND == False:
    print('Warning: Invalid ID')

```

**4 Write a function to drop a class**
* **Takes student ID number and class number**

```python
def drop_class(id,class_number):
  for d in student_database:
    if d['id']== id:
      if class_number not in d['courses']:    ## If typed in a wrong class number or the student did not register this class, it will print warning
        print('Warning: The class was not registered!')
      else:                                   ## Correct class number
        d['courses'].remove(class_number)
        print('The class is successfully removed!')
```

```python
### Test case 1: The student has registered this class
drop_class(87654,'EMSE6574')    
print(student_database)
```

```python
### Test case 2: Typed in a wrong class number
drop_class(34567,'ENG102')   
print(student_database)
```

```python
### Test case 3: id45678 did not register EMSE6574
drop_class(45678,'EMSE6574')  
print(student_database)
```
