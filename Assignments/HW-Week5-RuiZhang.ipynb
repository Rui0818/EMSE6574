13. Write a Pandas program to combine the columns of two potentially differently-indexed DataFrames into a single result DataFrame. 

```python
data3 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])

data4 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])
```

```python
data3
```

```python
data4
```

```python
# use join
data3.join(data4, how='outer')
```

```python
# use merge
pd.merge(data3, data4,left_index=True, right_index=True, how='outer')
```

```python
# use concat
pd.concat([data3,data4], axis=1,join='outer')
```
